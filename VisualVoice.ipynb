# pseudo-real-time test

```python
import numpy as np


def loudness(xs):
    return np.log10(np.sum(xs**2))


sr = 8000
full_audio = np.random.rand((16*8000))
full_audio_length = len(full_audio)

step_rate = 10            # steps per second
step = int(sr/step_rate)  # samples per second
max_step = 40             # maximum steps to feed into network
min_step = 10             # minimum steps to feed into network

audio = full_audio[:2*step]
last_power = loudness(audio[:step])
power = loudness(audio[-step:])
last_decreased = False
decreased = power-last_power < 0
step_count = 2
for idx in range(2*step,full_audio_length):
    if idx%step == 0:
        step_count += 1
        power = loudness(audio[-step:])
        # print(power)
        decreased = power-last_power < 0
        if (last_decreased and decreased and step_count >= min_step) or step_count >= max_step:
            # pause is detected if loudness decreased in the last two steps
            # print("pause")
            step_count = 0
            audio = np.array([])
        last_power = power
        last_decreased = decreased
    audio = np.append(audio,full_audio[idx])
```
